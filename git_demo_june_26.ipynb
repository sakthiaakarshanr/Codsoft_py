```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

print("First 5 rows of the Iris dataset:")
print(iris.head())

print("\nInformation about the dataset:")
iris.info()

print("\nDescriptive statistics of the dataset:")
print(iris.describe())

print("\nMissing values per column:")
print(iris.isnull().sum())

print("\nValue counts for the 'species' column:")
print(iris['species'].value_counts())

# Visualize the distribution of features and relationships
print("\nGenerating a pair plot for the Iris dataset...")
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset by Species', y=1.02) # Adjust title position
plt.show()

print("EDA complete. The pair plot shows relationships between features and species.")
```
